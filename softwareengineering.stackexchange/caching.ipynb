```python
%%sql
SELECT
    post_id,
    view_count,
    ntile(4) over (order by view_count) as quartile
FROM (SELECT * FROM actual_posts LIMIT 100) post_data
```

```python
%%sql
SELECT
    SUM(view_count),
    quartile
FROM (
    SELECT
        post_id,
        view_count,
        ntile(4) over (order by view_count) as quartile
    FROM (SELECT * FROM actual_posts LIMIT 100) post_data
) sum_data
GROUP BY quartile
```
